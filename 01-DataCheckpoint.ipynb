2.The Jeff Sackmann ATP Tennis dataset is a dataset containing records of professional men’s tennis matches played on the ATP tour. Each row in the dataset represents one completed match and includes information about players, match outcomes, and tournament details. It also includes a csv that details all given players personal statistics including but not limited to: height, age, handedness, win rate, and rounds played.

The key dependent variable we will be using in the dataset is match outcome, usually represented by winner and loser identifiers. A score is a series of points players have, with the first player reaching 6 being the winner. For example, the scores can be something like 6-2 or 6-4 depending on how each player does. Additionally, some matches on the dataset have scores such as 7-5, indicating a tie-breaker occurred in the set. A tie-breaker occurs when both players reach 5-5. We will be using scores to determine the match outcome, using binary code to identify wins and losses. For example, a 1 would signify a win while a 0 would signify a loss. These will act as our dependent variables in the study. 

The independent variables are players’ physical attributes and demographic characteristics, which we will be measuring to determine if these factors influence performance outcomes in professional tennis. These variables include height, handedness, and age. Each metric captures a distinct dimension of physical advantage or developmental stage that could affect win rate.

Height is recorded in centimeters (cm) in the ATP dataset. Professional male tennis players typically range from approximately 165 cm to over 210 cm in height. Height is especially relevant in tennis due to its impact on serving. Taller players can generate sharper serve angles and higher contact points, which greatly influence aces occurring in the match. An ace is when a player serves in a way that prevents the returner from even coming into contact with the ball. However, extreme height may reduce agility or endurance. With all of this being said, we want to analyze players to see if height acts as a benefit or drawback when playing tennis. If there is a benefit, what height makes being tall a “drawback” while playing tennis.

Handedness is an independent variable indicating whether a player is right-handed or left-handed. A majority of tennis players are right-handed, with left-handed players representing a minority (typically around 10-15% of the professional population). Handedness is a strategic key while playing tennis since most players are trained to play against right-handed players. Left-handed players often create unusual spin patterns and serve angles, which become especially challenging for players trained to expect the opposite. We will be measuring handedness through a binary variable (0 = right-handed, 1 = left-handed).

Age is calculated in years. Players generally enter the professional tennis league in their late teens, peak physically and competitively in their early-to-mid 20s, and experience performance decline in their 30s due to reduced recovery capacity and increased injury risk. With that being said, older players do have several advantages over younger players. For instance, older players have a plethora of experience and strategic intelligence, which is correlated with age. Therefore, age may exhibit a normal distribution with win rate rather than a strictly positive or negative one. These independent variables capture structural physical and biological characteristics that may influence match outcomes. These 3 independent variables will help signify the role physicality plays in tennis and if it affects match outcomes.
