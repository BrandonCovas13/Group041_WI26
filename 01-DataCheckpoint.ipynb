```python
## YOUR CODE TO LOAD/CLEAN/TIDY/WRANGLE THE DATA GOES HERE
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# A. Loading the Dataset:
mens_df = pd.read_csv("data/00-raw/atp_matches_2024.csv")

# B. Tidying the dataset:
# drop unneccessary columns
cols_to_drop = ['tourney_id', 'surface', 'draw_size', 'winner_seed', 'winner_entry', 'loser_seed', 'loser_entry', 'winner_ioc', 'loser_ioc']
mens_df = mens_df.drop(columns=cols_to_drop, axis=1)

# C. dataset size:
print('Size of the ATP mens datset: ', mens_df.shape, '\n')

# D find missing data:
#print(mens_df.isna().any())
# drop matches that are forfited:
mens_df = mens_df[mens_df['score'] != 'W/O']
# drop matches that were retired/RET (ended early for injury):
mens_df = mens_df[~mens_df['score'].str.contains('RET')]

# where is data missing at this point:
missing_info = mens_df.isnull().any(axis=1)
missing_info = mens_df[missing_info.eq(True)]
print('Tournaments where there are the most missing values:')
print(missing_info['tourney_name'].value_counts())
print('''From the above table, we can see that most null values for rank are from the Roland Garros, 
    \nParis Olympics, and Davis Cup tournament. We can conclude that there is some systematic issue 
    \nwith these specific tournaments and their data collection.\n\n''')

# F. clean data, drop the rows where player info is not recorded
# Since this information is critical to our analysis, we need data where all this information is intact
mens_df = mens_df.dropna()

# E find outliers:
# describe variables where it is possible to identify outliers
print('Description of variables to identify outliers:')
print(mens_df[['winner_hand', 'winner_ht', 'winner_age',
       'loser_id', 'loser_name', 'loser_hand', 'loser_ht', 'loser_age',
       'score', 'best_of', 'minutes', 'winner_rank', 'winner_rank_points',
       'loser_rank', 'loser_rank_points']].describe())

print('''\nFrom this table we can see that there ares some strange minimum values in some of the data. 
    \nWe will visualize it with boxplots: ''')

fig, axes = plt.subplots(1, 2, figsize=(8, 3))
axes[0].boxplot([mens_df['winner_age'], mens_df['loser_age']], labels=['Winner', 'Loser'])
axes[0].set_ylabel('Age')

axes[1].boxplot([mens_df['winner_ht'], mens_df['loser_ht']],labels=['Winner', 'Loser'])
axes[1].set_ylabel('Height (cm)')

plt.tight_layout()
plt.show()

fig, axes = plt.subplots(1, 2, figsize=(8, 3))
axes[0].boxplot(mens_df['minutes'], labels=['Game'])
axes[0].set_ylabel('minutes')

axes[1].boxplot([mens_df['winner_rank_points'], mens_df['loser_rank_points']],labels=['Winner', 'Loser'])
axes[1].set_ylabel('Rank Points')

plt.tight_layout()
plt.show()

# Remove player with outlier height
print('''From these plots we can identify a clear outlier in the loser heights at 71 cm. We should 
    \nremove this row if this player has an incorrectly recorded height.''')
print('Player(s) with outlier height: ', mens_df[mens_df['loser_ht'] < 100]['loser_name'])
print('After a Google search we can find that his height is not recorded on ATP, so we will drop this row from the table')
mens_df = mens_df[mens_df['loser_ht'] > 100]
plt.figure(figsize=(8,3))
plt.boxplot([mens_df['winner_ht'], mens_df['loser_ht']],labels=['Winner', 'Loser'])
plt.ylabel('Height (cm)')
print('Data without outlier:')
plt.show()
print('''The rest of this data seems like there is no clear outliers that are impossible to come accross in 
      \na tennis match. Although we see alot of outliers in the upper regions of the age and rank points chart, 
      \nthis doesn\'t cause alarm because of the structure of tennis ranking and the fact that many successful 
      \ntennis players continue to play later in life.''' )

# G. write data
mens_df.to_csv('data/02-processed/atp_matches_2024.csv')
```

```python
import pandas as pd

# A. Load data
wta = pd.read_csv('data/00-raw/wta_matches_2024.csv')

# B. Tidy 
# Select columns of interest
wta_data = wta[['winner_name', 'winner_hand', 'winner_ht', 'winner_age', 
                'loser_name', 'loser_hand', 'loser_ht', 'loser_age',
                'winner_rank', 'winner_rank_points', 'loser_rank', 'loser_rank_points']]

# C. Size of dataset
print(f"Size of the Women's Tennis Association (WTA) dataset: {wta_data.shape}\n")

# D. Missing data
print('Missing data per column:')
missing = wta_data.isnull().sum()
print(missing)
print('From the table above we can see that most of the missing data comes from the height measurements.')
print(f"There are {missing['winner_ht'] + missing['loser_ht']} missing height measurements, which we will have to exclude in order to conduct our analysis.\n")

# E. Outliers
print('Summary statistics of the data to identify outliers:')
print(wta_data.describe())
print(wta_data.describe(include='object'))

print('From this, we can see that there are 3 unique values for dominant hand preference.')
print(f"The unique values for the dominant hand preference are: {wta_data['winner_hand'].unique()}")
print(f"We want to look at strictly Right (R) or Left (L) handed players, and will remove outliers.\n")
ambidextrous_winners = wta_data[wta_data['winner_hand'] == 'U'].index
wta_data = wta_data.drop(ambidextrous_winners)

ambidextrous_losers = wta_data[wta_data['loser_hand'] == 'U'].index
wta_data = wta_data.drop(ambidextrous_losers)


# F. Clean data
# We are dropping all entries with missing values since as noted above, most missing data comes from the height measurements.
# Since height is a key biological factor in our analysis, we are only using entries including height data.
wta_data_clean = wta_data.dropna(how='any').reset_index()

# G. Write data
wta_data_clean.to_csv('data/02-processed/wta_matches_2024.csv')

# Look at data
print('Take a look at the cleaned data:')
wta_data_clean.head()
```
